# Conditional Selection:

## Single Condition Implementation:

### Question: To Get the name of freshers:

```python
table[table['Total Experience'] == 0]['Name']
```

### Question: To get the name, company and technology of the freshers:

```python
table[table['Total Experience'] == 0][['Name','Company','Technology']]
```

### Question: To get the entire details of the employee with the highest package:

```python
table[table['Salary Package '] == table['Salary Package '].max()]
```

### Question: To get the entire details of the employee that has the most experience:

```python
table[table["Total Experience"] == max(table['Total Experience'])]
```

### List the boolean series for which age > 30:

```python
table['Age']>30
```

## Double Condition Implementation with AND and OR operators:

### Question: To get the details of freshers whose age is greater than 27:

```python
table[(table['Total Experience'] == 0) & (table['Age'] > 27)]
```

### Question: To get the name, company and Technology of employees whose is a fresher or current experience is 0:

```python
table[(table['Total Experience'] == 0) | (table['Experience'] == 0)][['Name','Company','Technology']]
```
